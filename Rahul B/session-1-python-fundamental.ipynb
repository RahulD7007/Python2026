```python
# gusseing game
# generate a random integer between 1 and 100

import random
jackpot = random.randint(1,100)
guess = int(input('Please enter jackpot number'))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Wrong!! Please enter higher number')
  else:
    print('Wrong!! Please enter lower number')
  guess = int(input('Please enter jackpot number'))
  counter = counter + 1
else:
  print('Congratulation!! You are correct at ',counter,' round')
```

```python
# Linear search is an example of break
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
# Example
lower = int(input('Enter lower number: '))
upper = int(input('Enter upper number: '))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)
```
