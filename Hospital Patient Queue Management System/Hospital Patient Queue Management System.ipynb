## Hospital Patient Queue Management System:- 

### Introduction:-

- The Hospital Patient Queue Management System is a Python-based application designed to efficiently manage patient flow in a hospital. In real-world hospitals, handling multiple patients with different urgency levels can be challenging. This system provides a structured way to organize patients based on their condition (Normal, Urgent, and Super Urgent).

- The project uses Object-Oriented Programming (OOP) principles to model real-life entities such as patients and hospital departments (specializations). It allows users to add patients, view queues, serve patients based on priority, and remove patients when needed. This system helps simulate how hospitals prioritize critical cases while maintaining an organized workflow.

### Classes, Attributes & Methods:-

1. **Patient Class**:-

- The Patient class represents an individual patient in the hospital system.

- **Attributes**:-
    - name → Stores the name of the patient
    - status → Represents urgency level
        - 0 → Normal
        - 1 → Urgent
        - 2 → Super Urgent

- **Methods**:-
    - __init__(self, name, status) → Initializes patient details
    - __str__(self) → Returns formatted string of patient details
    - get_status_text(self) → Converts numeric status into readable format (Normal, Urgent, Super Urgent)

2. **Specialization Class**:-

- The Specialization class manages patient queues for different hospital departments.

- **Attributes**:-
    - name → Name of specialization (e.g., Cardiology, Neurology)
    - queue → List that stores patient objects
    - max_capacity → Maximum number of patients allowed in queue

- **Methods**:-
    - __init__(self, name, max_capacity=5) → Initializes specialization details
    - add_patient(self, patient) → Adds a patient based on priority → (Super Urgent → front, Urgent → middle, Normal → end)
    - get_next_patient(self) → Removes and returns the next patient (highest priority)
    - remove_patient(self, name) → Removes a patient by name
    - list_patients(self) → Displays all patients in the queue

3. **OperationsManager Class**:-

- The OperationsManager class controls the overall system and user interaction.

- **Attributes**:-
    - specializations → Dictionary storing multiple Specialization objects

- **Methods**:-
    - __init__(self) → Initializes different hospital departments
    - print_menu(self) → Displays menu options to the user
    - run(self) → Main loop to execute the program
    - add_patient(self) → Takes input and adds patient to selected specialization
    - list_patients(self) → Displays all patients from all specializations
    - get_next_patient(self) → Retrieves and displays next patient
    - remove_patient(self) → Removes a patient based on user input

### Key Concepts Used:- 

1. **Object-Oriented Programming (OOP)**:-
- The project is built using OOP concepts such as:-
    - Classes & Objects → Patient, Specialization, OperationsManager
    - Encapsulation → Data and methods grouped inside classes
    - Abstraction → Hides complex queue logic from the user

2. **Queue Data Structure**:- 
- The system uses a priority queue concept
- Patients are served based on urgency:-
    - Super Urgent → Highest priority
    - Urgent → Medium priority
    - Normal → Lowest priority

3. **Lists in Python**:-
    - Python lists are used to implement queues
    - Insert, delete, and traversal operations are performed

4. **Conditional Statements**:-
    - Used to check patient priority and manage queue positions

5. **User Interaction (CLI)**:-
    - Menu-driven program using input/output functions

## Programming Code for this Project:-

```python
# The Patient class represents an individual patient & includes 2 attributes name & status
class Patient:
    def __init__(self, name, status):
        self.name = name
        self.status = status  # 0 = normal, 1 = urgent, 2 = super urgent

    def __str__(self):
        return f"{self.name} ({self.get_status_text()})"

    def get_status_text(self):
        if self.status == 0:
            return "Normal"
        elif self.status == 1:
            return "Urgent"
        elif self.status == 2:
            return "Super Urgent"


# The Specialization class manages patient queues within different specializations.
class Specialization:
    def __init__(self, name, max_capacity=5):
        self.name = name
        self.queue = []
        self.max_capacity = max_capacity

    def add_patient(self, patient):
        if len(self.queue) >= self.max_capacity:
            print("Queue is full! Cannot add more patients.")
            return

        # Priority handling (higher status first)
        if patient.status == 2:
            self.queue.insert(0, patient)
        elif patient.status == 1:
            inserted = False
            for i in range(len(self.queue)):
                if self.queue[i].status < 1:
                    self.queue.insert(i, patient)
                    inserted = True
                    break
            if not inserted:
                self.queue.append(patient)
        else:
            self.queue.append(patient)

        print(f"Patient {patient.name} added successfully.")

    def get_next_patient(self):
        if not self.queue:
            print("No patients in queue.")
            return None
        return self.queue.pop(0)

    def remove_patient(self, name):
        for patient in self.queue:
            if patient.name.lower() == name.lower():
                self.queue.remove(patient)
                print(f"Patient {name} removed.")
                return
        print("Patient not found.")

    def list_patients(self):
        if not self.queue:
            print("No patients in this specialization.")
            return
        print(f"\nPatients in {self.name}:")
        for i, patient in enumerate(self.queue, start=1):
            print(f"{i}. {patient}")


# The OperationsManager class serves as the user interface for interacting with the Specialization instances.
class OperationsManager:
    def __init__(self):
        self.specializations = {
            1: Specialization("Cardiology"),
            2: Specialization("Neurology"),
            3: Specialization("Orthopedics"),
            4: Specialization("Pediatrics"),
        }

    def print_menu(self):
        print("\n===== Hospital Queue System =====")
        print("1. Add New Patient")
        print("2. List Patients")
        print("3. Get Next Patient")
        print("4. Remove Patient")
        print("5. Exit")

    def run(self):
        while True:
            self.print_menu()
            choice = input("Enter your choice: ")

            if choice == "1":
                self.add_patient()
            elif choice == "2":
                self.list_patients()
            elif choice == "3":
                self.get_next_patient()
            elif choice == "4":
                self.remove_patient()
            elif choice == "5":
                print("Exiting system. Thank you!")
                break
            else:
                print("Invalid choice!")

    # Adding new patients to specializations
    def add_patient(self):
        spec_id = int(input("Enter specialization (1-Cardio, 2-Neuro, 3-Ortho, 4-Pedia): "))
        name = input("Enter patient name: ")
        status = int(input("Enter status (0=Normal, 1=Urgent, 2=Super Urgent): "))

        if spec_id in self.specializations:
            patient = Patient(name, status)
            self.specializations[spec_id].add_patient(patient)
        else:
            print("Invalid specialization!")

    # Listing patients in specializations
    def list_patients(self):
        for spec in self.specializations.values():
            spec.list_patients()

    # Retrieving the next patient
    def get_next_patient(self):
        spec_id = int(input("Enter specialization: "))
        if spec_id in self.specializations:
            patient = self.specializations[spec_id].get_next_patient()
            if patient:
                print(f"Next patient: {patient}")
        else:
            print("Invalid specialization!")

    # Removing patients
    def remove_patient(self):
        spec_id = int(input("Enter specialization: "))
        name = input("Enter patient name to remove: ")

        if spec_id in self.specializations:
            self.specializations[spec_id].remove_patient(name)
        else:
            print("Invalid specialization!")


# Main Program
if __name__ == "__main__":
    system = OperationsManager()
    system.run()
```
