```python
class Employee:
    def __init__(self, name, salary):
        self.name = name # public attribute
        self.__salary = salary # private attribute
    def display(self): # public method
      print(self.name)
      print(self.__salary)



emp = Employee("Rahul",190000)
print(emp.name)
```

```python
class BankAccount:
  def __init__(self):
    self.balance=10000

  def __update_balnce(self,amount):
    self.balance+=amount

  def deposit(self,amount):
    self.__update_balnce(amount)

acc = BankAccount()
print(acc.balance)
acc.deposit(5000)
print(acc.balance)
acc.
```

```python
class Employee:
  def __init__(self):
    self.__salary = 50000
  def get_salary(self):
    return self.__salary
  def set_salary(self,salary):
    if type(salary) == int:
      self.__salary = self.__salary+salary
    else:
      print('salary must be integer')


emp = Employee()
print(emp.get_salary())
emp.set_salary(90000)
print(emp.get_salary())



```

```python
class Person:
  def __init__(self,name,country):
    self.name = name
    self.country = country

  def greet(self):
    if self.country == 'india':
      print('Namaste',self.name)
    else:
      print('Hello',self.name)
```
