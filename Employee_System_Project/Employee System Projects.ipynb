# Employees System Project:-

### Introduction:-

- The Employees System Project is a Python-based application that demonstrates the use of Object-Oriented Programming (OOP) concepts.

- It is designed to manage employee records efficiently by organizing data and operations into classes. The project includes three main classes:-

    - Employee
    - EmployeesManager
    - FrontendManager

- This system allows users to perform operations such as adding, viewing, updating, and deleting employee records.

### Classes:-

#### 1. Employee Class:-
- The Employee class represents a single employee.

- **Attributes**:-
    - name → Name of the employee.
    - age → Age of the employee.
    - salary → Salary of the employee.

- **Methods**:-
    - __str__() → Returns formatted employee details.
    - Used to display employee information in a readable format.

#### 2. EmployeesManager Class:-
- The EmployeesManager class manages all employee records using a list.

- **Functionalities**:-
    - Add a new employee
    - Display all employees
    - Delete employees within a given age range
    - Search for an employee by name
    - Update salary of an employee

#### 3. FrontendManager Class:- 
- The FrontendManager class acts as the user interface.

- **Responsibilities**:-
    - Takes user input
    - Calls appropriate methods from EmployeesManager
    - Displays results to the user

- **Features**:-
    - Add employee details
    - View employee list
    - Delete employees by age range
    - Update employee salary

### Key Concepts Used:- 
- Object-Oriented Programming (OOP)
- Classes and Objects
- Encapsulation
- Method calling between classes
- User interaction (input/output)

### Advantages of this Project:-
- Simple and easy to understand
- Demonstrates real-world use of OOP
- Modular and reusable code
- Easy to extend (e.g., add database, GUI)

### Conclusion:- 

The Employees System Project is a basic yet effective example of applying OOP concepts in Python. It helps in understanding how different classes interact to manage and manipulate data efficiently.

## Programming Code for this Project:-

```python
# To organize code into reusable and manageable components, OoP's concept is used in this project.
# Employee Class
class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Salary: {self.salary}"

# Role of EmployeesManager Class -> For handling all employee data operations. 
# EmployeesManager Class
class EmployeesManager:
    def __init__(self):
        self.employees = []

    # Add a new employee to the list.
    def add_employee(self, name, age, salary):
        emp = Employee(name, age, salary)
        self.employees.append(emp)
        print("Employee added successfully!")

    # List all existing employees.
    def list_employees(self):
        if not self.employees:
            print("No employees found.")
        else:
            for emp in self.employees:
                print(emp)

    # Delete employees within a specified age range.
    def delete_by_age(self, min_age, max_age):
        original_count = len(self.employees)
        self.employees = [
            emp for emp in self.employees
            if not (min_age <= emp.age <= max_age)
        ]
        deleted = original_count - len(self.employees)
        print(f"{deleted} employee(s) deleted.")

    # Find an employee by their name.
    def find_employee(self, name):
        for emp in self.employees:
            if emp.name.lower() == name.lower():
                print("Employee found:")
                print(emp)
                return emp
        print("Employee not found.")
        return None

    # Update an employee's salary by name.
    def update_salary(self, name, new_salary):
        emp = self.find_employee(name)
        if emp:
            emp.salary = new_salary
            print("Salary updated successfully!")


# FrontendManager Class provides a User Interface for interacting with the EmployeeManager.
class FrontendManager:
    def __init__(self):
        self.manager = EmployeesManager()

    def menu(self):
        while True:
            print("\n===== Employee Management System =====")
            print("1. Add Employee")
            print("2. List Employees")
            print("3. Delete Employees by Age Range")
            print("4. Find Employee by Name")
            print("5. Update Salary")
            print("6. Exit")

            choice = input("Enter your choice: ")

            # For adding new employees:-
            if choice == "1":
                name = input("Enter name: ")
                age = int(input("Enter age: "))
                salary = float(input("Enter salary: "))
                self.manager.add_employee(name, age, salary)

            # For Listing existing employees:-
            elif choice == "2":
                self.manager.list_employees()

            # For Deleting employees based on age range.
            elif choice == "3":
                min_age = int(input("Enter minimum age: "))
                max_age = int(input("Enter maximum age: "))
                self.manager.delete_by_age(min_age, max_age)

            # For finding employees by using name:-
            elif choice == "4":
                name = input("Enter name to search: ")
                self.manager.find_employee(name)

            # For updating employee salaries by using name:-
            elif choice == "5":
                name = input("Enter name: ")
                new_salary = float(input("Enter new salary: "))
                self.manager.update_salary(name, new_salary)

            elif choice == "6":
                print("Exiting program...")
                break

            else:
                print("Invalid choice! Try again.")


# Run the program
if __name__ == "__main__":
    app = FrontendManager()
    app.menu()
```
