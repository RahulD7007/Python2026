```python
# Airthematic Operators
print(5+6)
print(5-6)
print(5*6)
print(5/2)
print(5//2)
print(5%2)
print(5**2)
```

```python
# Relation Operators
print(4>5)
print(4<5)
print(4>=5)
print(4>=5)
print(4==4)
print(4!=4)
```

```python
# Logical Operators
print(1 and 0)
print(1 or 0)
print(not 4)
```

```python
# Bitwise Operator
print(2 & 3) # Bitwise and
# 2 -> 010
# 3 -> 110
# -------------
# 2 - >    010

print(2 | 3) # Bitwise or

print(2 ^ 3) # Bitwise XOR

print(~3)

print(4>>2)
print(4<<2)
```

```python
# Assignment operator => =

a = 2
#a = a+2
a+=2
print(a)
# ++a or a++

b = 10
#b = b + 5
print(b)
b+=5
print(b)


```

```python
# Membership operator
# in / not in
print('P' in ('Pune'))
print('P' not in ('Pune'))
print(2 in [2,3,4,5,6])
```
