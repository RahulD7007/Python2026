```python
# Creating String

s = 'hello'
print(s)
s = "hello"
print(s)
# Multiline string
s = '''hello'''
print(s)
s = """hello"""
print(s)
s = str('hello')
print(s)
```

```python
# why single and double quato in python
#s = 'Its raining outside'
#print(s)
s = 'It\'s raining outside'
print(s)
s = "It's raining outside"
print(s)
```

```python
#Accessing String
s = 'hello world'

#Indexing
print(s[6])
print(s[0])

# Negative indexing
print(s[-1])
```

```python
# Slicing
# Positive Index
s = 'hello world'
print(s[0:5])
print(s[2:3])
print(s[2:])
print(s[:3])
print(s[:])
print(s[0:6:2])
```

```python
# Negative Indexing
s = 'hello world'
print(s[6:0:-1])

# Reverse string
print(s[::-1])

s = 'hello world'
print(s[-5:])

print(s[-1:-6:-1])
print(s[-1:-6:-2])
```

```python
# Editing & Ddeleting string

# s = 'hello world'
# s[0]='H'

# Python strings are immutable

s = 'hello world'
del s

s = 'hello world'

del s[-1:-5:-2]
print(s)
```

```python
# Airthmetic operation on string

print("Pune" + "Mumbai")
print("Pune" +' '+ "Mumbai")

print('pune'*5)

print("*" *50)
```

```python
# Relational operators
'pune'== 'mumbai'
'pune'=='pune'
'pune'=='Pune'
'pune'!='mumbai'
'pune'>'mumbai'
```

```python
# String functions

print(len('hello world'))
print(max('hello world'))
print(min('hello world'))
print(sorted('hello world')) # return should be list
print(sorted('hello world',reverse=True))
```

```python
print(ord('P'))
```

```python
# capitalize/Title/Upper/Lower/Swapcase

s = 'Hello World'
print(s.capitalize())
print(s.title())
print(s.upper())
print(s.lower())
print(s.swapcase())
```

```python
# count/find/Index
# Index is similer to find function except if given character is not present in string it will throw error
s = 'My name is rahul'
print(s.count('a'))
print(s.find('a'))
print(s.find('z'))
print(s.index('a'))
print(s.index('z'))
```

```python
# endwith/startwith
s = 'My name is rahul'
print(s.endswith('l'))
print(s.endswith('rahul'))
print(s.endswith('p'))
print(s.endswith('uuuu'))

print(s.startswith('M'))
print(s.startswith('My'))
print(s.startswith('m'))
```

```python
# format
name = 'rahul'
gender = 'male'
print('Hi My name is {} I am a {}'.format(name,gender))
print('Hi My name is {1} I am a {0}'.format(name,gender))
print('Hi My name is {0} I am a {1}'.format(name,gender))
print('Hi My name is {} I am a {}'.format(gender,name))
```
