### Python Challenge 01: Student Marks Analyzer:-

- Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
- Your program should consist of:-
    1. Let the user input multiple students with their marks (name + integer score).
    2. After input is complete, display:
       - Average marks
       - Highest marks and student(s) who scored it
       - Lowest marks and student(s) who scored it
       - Total number of students

### Bonus:-
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names


```python
def input_students():
    students = {}

    print("Enter student details (type 'done' to finish):\n")

    while True:
        name = input("Enter student name: ").strip()

        if name.lower() == "done":
            break

        if name in students:
            print("Student already exists! Try a different name.\n")
            continue

        try:
            marks = int(input("Enter marks: "))
            students[name] = marks
            print("Added successfully!\n")
        except ValueError:
            print("Invalid marks! Please enter an integer.\n")

    return students


def analyze_students(students):
    if not students:
        print("No data to analyze.")
        return

    marks_list = list(students.values())

    total_students = len(students)
    average_marks = sum(marks_list) / total_students
    highest_marks = max(marks_list)
    lowest_marks = min(marks_list)

    # Students with highest marks
    top_students = [name for name, marks in students.items() if marks == highest_marks]

    # Students with lowest marks
    low_students = [name for name, marks in students.items() if marks == lowest_marks]

    # Report
    print("\n" + "=" * 40)
    print("STUDENT MARKS REPORT")
    print("=" * 40)

    print(f"Total Students   : {total_students}")
    print(f"Average Marks    : {average_marks:.2f}")
    print(f"Highest Marks    : {highest_marks} ({', '.join(top_students)})")
    print(f"Lowest Marks     : {lowest_marks} ({', '.join(low_students)})")

    print("=" * 40)


def main():
    students = input_students()

    while True:
        print("\nOptions:")
        print("1. View Report")
        print("2. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            analyze_students(students)
        elif choice == "2":
            print("Exiting program...")
            break
        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

### Python Challenge 02:- CLI Contact Book (CSV-Powered)

- Create a terminal-based contact book tool that stores and manages contacts using a CSV file.

- Your program should consist of:-
    1. Ask the user to choose one of the following options:
       - Add a new contact
       - View all contacts
       - Search for a contact by name
       - Exit
    2. Store contacts in a file called `contacts.csv` with columns:
       - Name
       - Phone
       - Email
    3. If the file doesn't exist, create it automatically.
    4. Keep the interface clean and clear.

### Example:-
- Add Contact
- View All Contacts
- Search Contact
- Exit

### Bonus:-
- Format the contact list in a table-like view
- Allow partial match search
- Prevent duplicate names from being added


```python
import csv
import os

FILE_NAME = "contacts.csv"


# Ensure file exists
def initialize_file():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Name", "Phone", "Email"])


# Add contact
def add_contact():
    name = input("Enter Name: ").strip()

    # Check duplicate
    with open(FILE_NAME, "r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["Name"].lower() == name.lower():
                print("Contact already exists!\n")
                return

    phone = input("Enter Phone: ").strip()
    email = input("Enter Email: ").strip()

    with open(FILE_NAME, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([name, phone, email])

    print("Contact added successfully!\n")


# View all contacts
def view_contacts():
    with open(FILE_NAME, "r") as file:
        reader = list(csv.reader(file))

        if len(reader) <= 1:
            print("No contacts found.\n")
            return

        print("\nCONTACT LIST")
        print("-" * 50)
        print(f"{'Name':<15}{'Phone':<15}{'Email':<20}")
        print("-" * 50)

        for row in reader[1:]:
            print(f"{row[0]:<15}{row[1]:<15}{row[2]:<20}")

        print("-" * 50 + "\n")


# Search contact (partial match)
def search_contact():
    search_name = input("Enter name to search: ").strip().lower()
    found = False

    with open(FILE_NAME, "r") as file:
        reader = csv.DictReader(file)

        print("\nSEARCH RESULTS")
        print("-" * 50)
        print(f"{'Name':<15}{'Phone':<15}{'Email':<20}")
        print("-" * 50)

        for row in reader:
            if search_name in row["Name"].lower():
                print(f"{row['Name']:<15}{row['Phone']:<15}{row['Email']:<20}")
                found = True

        if not found:
            print("No matching contact found.")

        print("-" * 50 + "\n")


# Main menu
def main():
    initialize_file()

    while True:
        print("====== CONTACT BOOK ======")
        print("1. Add Contact")
        print("2. View All Contacts")
        print("3. Search Contact")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_contact()
        elif choice == "2":
            view_contacts()
        elif choice == "3":
            search_contact()
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice!\n")


if __name__ == "__main__":
    main()
```

### Python Challenge 03:- Personal Movie Tracker with JSON

- Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.

- Your program should consist of:-
    1. Store all movie data in a 'movies.json' file.
    2. Each movie should have:
       - Title
       - Genre
       - Rating (out of 10)
    3. Allow the user to:
       - Add a movie
       - View all movies
       - Search movies by title or genre
       - Exit the app

### Bonus:-
- Prevent duplicate titles from being added
- Format output in a clean table
- Use JSON for reading/writing structured data


```python
import json
import os

FILE_NAME = "movies.json"


# Ensure file exists
def initialize_file():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w") as file:
            json.dump([], file)


# Load movies
def load_movies():
    with open(FILE_NAME, "r") as file:
        return json.load(file)


# Save movies
def save_movies(movies):
    with open(FILE_NAME, "w") as file:
        json.dump(movies, file, indent=4)


# Add movie
def add_movie():
    movies = load_movies()

    title = input("Enter movie title: ").strip()

    # Prevent duplicate titles
    for movie in movies:
        if movie["title"].lower() == title.lower():
            print("Movie already exists!\n")
            return

    genre = input("Enter genre: ").strip()

    try:
        rating = float(input("Enter rating (0-10): "))
        if rating < 0 or rating > 10:
            raise ValueError
    except ValueError:
        print("Invalid rating! Must be between 0 and 10.\n")
        return

    movies.append({
        "title": title,
        "genre": genre,
        "rating": rating
    })

    save_movies(movies)
    print("Movie added successfully!\n")


# View all movies
def view_movies():
    movies = load_movies()

    if not movies:
        print("No movies found.\n")
        return

    print("\nMOVIE LIST")
    print("-" * 50)
    print(f"{'Title':<20}{'Genre':<15}{'Rating':<10}")
    print("-" * 50)

    for movie in movies:
        print(f"{movie['title']:<20}{movie['genre']:<15}{movie['rating']:<10}")

    print("-" * 50 + "\n")


# Search movies (by title or genre)
def search_movies():
    query = input("Enter title or genre to search: ").strip().lower()
    movies = load_movies()
    found = False

    print("\nSEARCH RESULTS")
    print("-" * 50)
    print(f"{'Title':<20}{'Genre':<15}{'Rating':<10}")
    print("-" * 50)

    for movie in movies:
        if query in movie["title"].lower() or query in movie["genre"].lower():
            print(f"{movie['title']:<20}{movie['genre']:<15}{movie['rating']:<10}")
            found = True

    if not found:
        print("No matching movies found.")

    print("-" * 50 + "\n")


# Main menu
def main():
    initialize_file()

    while True:
        print("====== MOVIE TRACKER ======")
        print("1. Add Movie")
        print("2. View All Movies")
        print("3. Search Movie")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_movie()
        elif choice == "2":
            view_movies()
        elif choice == "3":
            search_movies()
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice!\n")


if __name__ == "__main__":
    main()
```
